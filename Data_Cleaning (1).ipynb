## Data Cleaning

```python
import pandas as pd

# Load the CSV file located in the same directory as the notebook
df = pd.read_csv("hotel_bookings.csv")

# Preview the first few rows
df.head()
```

```python
#Check the data types
df.info()
```

```python
#Check the data types
df.info()
```

```python
# Check for duplicate rows
duplicates = df.duplicated()
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
#determine columns with missing values
sum_missing_values = df.isna().sum()
print(sum_missing_values)
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
#check if missing values were dropped
sum_missing_values = df.isna().sum()
print(sum_missing_values)
```

```python
# Drop 'agent' and 'company' columns
df = df.drop(columns=['agent', 'company'])

# Confirm changes
print(f"Shape after cleaning: {df.shape}")
df.head()
```

```python
sum_missing_values = df.isna().sum()
print(sum_missing_values)
```

```python
# Count how many rows have 'Undefined' in market_segment
undefined_count = df[df['market_segment'] == 'Undefined'].shape[0]
print(f"Number of 'Undefined' rows in market_segment: {undefined_count}")
```

```python
# Drop rows where market_segment is 'Undefined'
df = df[df['market_segment'] != 'Undefined']
```

```python
# Confirm that the rows were dropped
undefined_count = df[df['market_segment'] == 'Undefined'].shape[0]
print(f"Number of 'Undefined' rows in market_segment: {undefined_count}")
```

```python
# Create datetime column
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d'
)
 
# Derived features
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = (
    df['adults'] + df['children'].fillna(0) + df['babies']
).astype(int)
```

```python
# Display the first 5 rows of the new columns
print(df[['arrival_date', 'total_nights', 'total_guests']].head())
```
